```python
df = df[df["Year"] >= 2011]
df = df[df["Year"] <= 2015]
```

```python
df_sum_console_Glob = df.groupby(["Console"])["Global_Sales"].agg('sum').reset_index()
# df_popl_NA = df[df["NA_Sales"] >= 300]
df_sum_console_Glob = df_sum_console_Glob.sort_values(by = ["Global_Sales"], ascending = False)
# df_sum_publisher_NA = df_sum_publisher_NA[df_sum_publisher_NA["NA_Sales"] >= 100]
plt.figure(figsize=(10,5))
chart = sns.barplot(
    data=df_sum_console_Glob,
    x='Console',
    y = 'Global_Sales',
    palette='Set1'
)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
chart.set_title("Number of Games by Console")
```

```python
plt.figure(figsize=(10,5))
chart = sns.countplot(
    data=df,
    x='Year',
    palette='Set1'
)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
chart.set_title("Number of Games by Year")
```
